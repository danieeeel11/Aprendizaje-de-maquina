```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
```

```python
X = data.data
```

```python
y = data.target
y = keras.utils.to_categorical(y)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(10, input_dim=X.shape[1], activation='relu'))
model.add(keras.layers.Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs=2000, batch_size=32, verbose=1, validation_split=0.3)
history
```

```python
plt.plot(history.history['loss'], label='loss - train')
plt.plot(history.history['val_loss'], label='loss - test')
plt.title('Loss Train vs Test')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show
```

```python
plt.plot(history.history['accuracy'], label='accuracy - train')
plt.plot(history.history['val_accuracy'], label='accuracy - test')
plt.title('Accuracy Train vs Test')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show
```
