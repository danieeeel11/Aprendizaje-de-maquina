```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier, SGDRegressor, LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import export_text
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv')
df.head()
```

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

Neurona

```python
model = keras.models.Sequential()

model.add(keras.layers.Input(shape=8))
#Capa 1
model.add(keras.layers.Dense(64, activation='relu'))
#Capa 2
model.add(keras.layers.Dense(64, activation='relu'))
#Capa 3
model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='sgd', 
              loss=keras.losses.binary_crossentropy, 
              metrics='accuracy')
```

```python
history = model.fit(X_scaled, y, epochs=1000, validation_split=0.3)
history
```

```python
#max_accuracy = max(history.history['accuracy'])
#print('Max Accuracy:', max_accuracy)
```

```python
valor_train = history.history['accuracy'][len(history.history['accuracy'])-1]
valor_test = history.history['val_accuracy'][len(history.history['val_accuracy'])-1]
print("Despues de entrenar el modelo de neuronas:")
print("  * Valor train (accuracy) =", round(valor_train*100, 2), "%")
print("  * Valor test (val_accuracy) =", round(valor_test*100, 2), "%")
```
