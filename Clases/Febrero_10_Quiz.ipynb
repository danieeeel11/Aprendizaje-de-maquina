```python
import pandas  as pd
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
import math
```

```python
df = pd.DataFrame([[1,2.1],
                  [2,3.9],
                  [3,6.1],
                  [4,8.2],
                  [5,10]], columns = ["x","y"])
```

```python
df
```

```python
X = df[["x"]]
y = df["y"]
```

```python
model = DecisionTreeRegressor()
model.fit(X,y)
```

Arbol hasta segundo nivel

```python
plt.figure(figsize=(15,10))
plot_tree(model,feature_names=model.feature_names_in_,class_names=["0", "1"], max_depth=1)
plt.show()
```

```python
df_t = pd.DataFrame([[1,2.1],
                  [2,3.9],
                  [3,6.1],
                  [4,8.2],
                  [5,10]], columns = ["x","y"])
```

```python
y_1 = df_t["y"].iloc[0:3].sum()/df_t["y"].iloc[0:3].count()
print(y_1)
```

```python
y_2 = df_t["y"].iloc[3:].sum()/df_t["y"].iloc[3:].count()
print(y_2)
```

```python
x=0
for i in range(3):
  x = x + pow(df_t["y"].iloc[i] - y_1,2)
sum_y1 = x
print(sum_y1)
```

```python
x=0
for i in range(2):
  x = x + pow(df_t["y"].iloc[i+3] - y_2,2)
sum_y2 = x
print(sum_y2)
```

```python
mse_si = 1/df_t["y"].iloc[0:3].count() * (sum_y1)
```

```python
mse_no = 1/df_t["y"].iloc[3:].count() * (sum_y2)
```

```python
mse_total = 3/5 * mse_si + 2/5 * mse_no
print(mse_total)
```
