```python
import pandas  as pd
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame([[1,2.1],
                  [2,3.9],
                  [3,6.1],
                  [4,8.2],
                  [5,10]], columns = ["x","y"])
```

```python
df
```

```python
X = df[["x"]]
y = df["y"]
```

```python
model = DecisionTreeRegressor()
model.fit(X,y)
```

Dibujo arbol de entropia

```python
plt.figure(figsize=(15,10))
plot_tree(model,feature_names=model.feature_names_in_,class_names=["0", "1"])
plt.show()
```
