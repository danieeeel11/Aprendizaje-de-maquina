```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
```

```python
data = load_breast_cancer()
```

```python
X = data.data
y = data.target
```

```python
cls = RandomForestClassifier(n_estimators=3, max_depth=2)
```

```python
cls.fit(X,y)
```

```python
muestra = X[0:1,:]
```

```python
cls.predict(muestra)
```

```python
cls.predict_proba(muestra)
```

```python
cls.estimators_
```

# Dibujar cada uno de los árboles

```python
plt.figure(figsize=(10,5))
plot_tree(cls.estimators_[0])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plot_tree(cls.estimators_[1])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plot_tree(cls.estimators_[2])
plt.show()
```
