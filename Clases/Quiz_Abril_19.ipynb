```python
import matplotlib.pyplot as plt
```

```python
x1 = [0,1,0,1]
x2 = [0,0,1,1]
y =  [0,1,1,0]
```

```python
for i in range(len(y)):
  if y[i] == 0:
    plt.plot(x1[i],x2[i], "o", color = "red")
  else:
    plt.plot(x1[i],x2[i], "o", color = "blue")
plt.show()
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

data = list(zip(x1, x2))
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(data, y)
```

```python
a_predecir = [[0,0.5],[0.5,1],[1,0.2]]
```

```python
predict = knn.predict(a_predecir)
```

```python
for i in range(len(a_predecir)):
  print("Punto ", a_predecir[i], " clase: ", predict[i])
```

```python
for i in range(len(y)):
  if y[i] == 0:
    plt.plot(x1[i],x2[i], "o", color = "red")
  else:
    plt.plot(x1[i],x2[i], "o", color = "blue")
for i in range(len(predict)):
  if predict[i] == 0:
    plt.plot(a_predecir[i][0],a_predecir[i][1], "o", color = "red")
  else:
    plt.plot(a_predecir[i][0],a_predecir[i][1], "o", color = "blue")
plt.show()
```
