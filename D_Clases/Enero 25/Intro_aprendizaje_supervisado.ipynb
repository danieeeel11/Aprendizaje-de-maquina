# Determinar si un numero x es un numero positivo o no

##Enfoque tradicional

```python
def predecir(x):
  if x>0:
    return True
  else: 
    return False
```

```python
predecir(-6)
```

##Enfoque Machine Learning - Aprendizaje supervisado

```python
#Ejemplos
X =  [[-3], [-2], [-20], [-0.1], [0.1], [0.5], [10], [15]]
y = ["NO POSITIVO", "NO POSITIVO", "NO POSITIVO", "NO POSITIVO", "POSITIVO", "POSITIVO", "POSITIVO", "POSITIVO",]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo = DecisionTreeClassifier()
```

Entrenar el sistema

```python
modelo.fit(X,y)
```

```python
modelo.predict([[0.001]])
```

```python
from sklearn.tree import plot_tree
```

Impresion arbol de decision

```python
plot_tree(modelo)
```


# Determinar si un numero x es un numero positivo, negativo o neutro

## Enfoque tradicional

```python
def predecir(x):
  if x>0:
    return True
  elif x<0:
    return False
  else: 
    return 'Es cero'
```

```python
print(predecir(-6))
print(predecir(0))
print(predecir(10))
```

## Enfoque Machine Learning - Aprendizaje supervisado

```python
#Ejemplos
X =  [[-3], [-2], [-20], [-0.1], [-0.00000000001], [0], [0.00000000001], [0.1], [0.5], [10], [15]]
y = ["NO POSITIVO", "NO POSITIVO", "NO POSITIVO", "NO POSITIVO", "NO POSITIVO", "NEUTRO", "POSITIVO", "POSITIVO", "POSITIVO", "POSITIVO", "POSITIVO",]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo = DecisionTreeClassifier()
```

Entrenar el sistema

```python
modelo.fit(X,y)
```

```python
print(modelo.predict([[0.1]]))
print(modelo.predict([[-0.1]]))
print(modelo.predict([[0]]))
```

```python
from sklearn.tree import plot_tree
```

Impresion arbol de decision

```python
plot_tree(modelo)
```
